```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('medical_examination.csv')
df.head()
```

```python
def overweight(row):
    bmi = row['weight']/((row['height']/100)**2)
    if bmi>25:
        return 1
    else:
        return 0
df['overweight'] = df.apply(lambda row : overweight(row), axis= 1)
df.head()
```

```python
df['cholesterol'] = np.where(df['cholesterol'] == 1, 0, 1)
df['gluc'] = np.where(df['gluc'] == 1, 0, 1)
df.head()
```

```python
df_cat = pd.melt(df, id_vars= 'cardio', value_vars=['active', 'alco', 'cholesterol', 'gluc', 'overweight', 'smoke'])
df_cat
```

```python
df_cat = pd.DataFrame(df_cat.groupby(['cardio', 'variable', 'value'])['value'].count())
df_cat.rename(columns={'value':'total'}, inplace= True)
df_cat.reset_index(inplace=True)
df_cat
```

```python
graph = sns.catplot(x='variable', y='total', hue='value', col='cardio', data=df_cat, kind='bar')
fig = graph.fig
```

```python
df_heat = df.loc[(df['ap_lo'] <= df['ap_hi']) & (df['height'] >= df['height'].quantile(0.025)) & (df['height'] <= df['height'].quantile(0.975)) & (df['weight'] >= df['weight'].quantile(0.025)) & (df['weight'] <= df['weight'].quantile(0.975))]
df_heat.head()
```

```python
corr = df_heat.corr()
corr
```

```python
mask = np.triu(np.ones_like(corr, dtype=bool))
mask
```

```python
fig, ax = plt.subplots(figsize=(12,12))
sns.heatmap(ax=ax, data=corr, mask=mask, annot=True, fmt='.1f')
```
